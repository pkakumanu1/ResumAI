```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
BR_df = pd.read_csv("Building_Permits.csv")
CP_df = pd.read_csv("Car_Production.csv")
COP_df = pd.read_csv("Crude_Oil_Production.csv")
DG_df = pd.read_csv("Durable_Goods.csv")
HS_df = pd.read_csv("House_Starts.csv")
UR_df = pd.read_csv("Unemployment_Rate.csv")
```

```python
UR_df.rename(columns = {'value':'UnemploymentRate'}, inplace = True)
BR_df.rename(columns = {'value':'BuildingPermits'}, inplace = True)
CP_df.rename(columns = {'value':'CarProduction'}, inplace = True)
COP_df.rename(columns = {'value':'CrudeOilProduction'}, inplace = True)
DG_df.rename(columns = {'value':'DurableGoods'}, inplace = True)
HS_df.rename(columns = {'value':'HouseStarts'}, inplace = True)
```

```python
UR_df['country'] = UR_df['description'].apply(lambda  x : x.split(":")[0]  )
BR_df['country'] = BR_df['description'].apply(lambda  x : x.split(":")[0]  )
CP_df['country'] = CP_df['description'].apply(lambda  x : x.split(":")[0]  )
COP_df['country'] = COP_df['description'].apply(lambda  x : x.split(":")[0])
DG_df['country'] = DG_df['description'].apply(lambda  x : x.split(":")[0]  )
HS_df['country'] = HS_df['description'].apply(lambda  x : x.split(":")[0]  )
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_squared_log_error
```

```python
X = result.drop("UnemploymentRate", axis=1)
y = result["UnemploymentRate"]
print(X.shape, y.shape)
```

```python
# Split the data using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
import numpy as np

model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
from matplotlib import pyplot
importance = model.coef_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```
